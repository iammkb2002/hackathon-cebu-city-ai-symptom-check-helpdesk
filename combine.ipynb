````python
import os

def is_ignored(path):
    return 'node_modules' in path.split(os.sep) or '.next' in path.split(os.sep) or '.git' in path.split(os.sep) or path in ['.gitattributes', 'combine.ipynb', 'package-lock.json', '.env.local']

def get_all_files():
    all_files = []
    for root, dirs, files in os.walk('.'):
        for file in files:
            filepath = os.path.join(root, file)
            relative_path = os.path.relpath(filepath, start='.')
            if not is_ignored(relative_path):
                all_files.append(relative_path)
    return all_files

def read_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            return f.read()
    except UnicodeDecodeError:
        with open(file_path, 'r', encoding='latin-1') as f:
            return f.read()

def compile_files():
    file_list = get_all_files()
    file_structure = "File Structure:\n"
    compiled_content = ""
    for file in file_list:
        file_structure += f"{file}\n"
        code = read_file(file)
        compiled_content += f"{file}\n```\n{code}\n```\n\n"
    return file_structure + "\n" + compiled_content

compiled_content = compile_files()

# Display the compiled content
print(compiled_content)
````
